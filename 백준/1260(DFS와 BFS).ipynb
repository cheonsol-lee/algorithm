```python
N, M, V = map(int, input().split())
```

```python
graph = [[] for i in range(N + 1)]
```

```python
graph
```

```python
for i in range(M):
    n1, n2 = map(int, input().split())
    graph[n1].append(n2)
    graph[n2].append(n1)
```

```python
for i in range(1, N+1):
    graph[i].sort()
```

```python
visited_dfs = [False for i in range(N + 1)]
```

```python
visited_dfs
```

```python
def dfs(graph, v, visited):
    visited[v] = True
    print(v, end=' ')
    
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
```

```python
dfs(graph, V, visited_dfs)
```

```python
visited_bfs = [False for i in range(N + 1)]
```

```python
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    print(start, end=' ')
    
    while queue:
        v = queue.popleft()
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
                print(i, end=' ')
```

```python
bfs(graph, V, visited_bfs)
```
