```python
# 상하좌우, 대각선, 상하좌우 2칸씩 띄기
ban_condition_x = [0, -1,-1,-1,0,0,1,1,1, -2,0,0,2]
ban_condition_y = [0, -1,0,1,-1,1,-1,0,1, 0,-2,2,0]
```

```python
def ban_confirm(seed_1, seed_2):
    x = seed_1[0]
    y = seed_1[1]
    
    for z in range(13):
        nx = x + ban_condition_x[z]
        ny = y + ban_condition_y[z]
#         print((nx,ny))
        if (nx,ny) == seed_2:
            return False # 하나라도 충돌이 있으면 False
        else:
            continue
            
    return True # 모두 충돌하지 않는다면 True
```

```python
seed_1 = (1,1)
seed_2 = (1,4)
seed_3 = (3,1)
print(ban_confirm(seed_1, seed_2))
print(ban_confirm(seed_1, seed_3))
print(ban_confirm(seed_2, seed_3))
if(ban_confirm(seed_1, seed_2) and ban_confirm(seed_1, seed_3) and ban_confirm(seed_2, seed_3)):
    print('통과')
```

```python
length = len(seed_list)
min_cost = 987654321

for i in range(0, length):
    for j in range(i+1, length):
        for k in range(j+1, length):
            # i,j,k 가 가능한 조합인지 파악
            seed_1 = seed_list[i]
            seed_2 = seed_list[j]
            seed_3 = seed_list[k]
            
            # 모두 True 일 때, 3가지 씨앗이 가능한 경우이므로 최소 비용 저장
            if(ban_confirm(seed_1, seed_2) and ban_confirm(seed_1, seed_3) and ban_confirm(seed_2, seed_3)):
                cost = sum_list[i] + sum_list[j] + sum_list[k]

                # 최소 비용 저장
                if cost < min_cost:
                    min_cost = cost
                    print(f'i:{i}, j:{j}, k:{k} cost:{min_cost}')
```

```python
n = int(input())
maps = [list(map(int, input().split())) for _ in range(n)]
directions = [(0, 0), (0, 1), (0, -1), (1, 0), (-1, 0)]
```

```python
def check(lst):
# i, j, k의 땅값을 리턴
    ret = 0
    locations = []
    
    for flower in lst:
        x = flower // n # 행
        y = flower % n # 열
        if x == 0 or x == n - 1 or y == 0 or y == n - 1:
            return 10000
            
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            ret += maps[nx][ny]
            locations.append((nx, ny))
            
    if len(set(locations)) != 15:
        return 10000
    
    print(len(locations))
    return ret
    
```

```python
ans = 100000
for i in range(n * n):
    for j in range(i + 1, n * n):
        for k in range(j + 1, n * n):
            ans = min(ans, check([i, j, k]))
            print(f'i:{i}, j:{j}, k:{k} cost:{ans}')
              
```
